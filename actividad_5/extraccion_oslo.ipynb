```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python

data=pd.read_csv('listings_50_outliers_oslo.csv')
```

```python
data.info()
```

```python
freq_tbl(data)
```

```python
table1 = freq_tbl(data[ 'host_name'])
table1
```

```python
table2=table1.drop(['percentage', 'cumulative_perc'], axis=1)
table2
```

```python

filtro = table2[table2['frequency'] > 60]

filtro
```

```python
filtro_index=filtro.set_index('host_name')
filtro_index
```

```python
filtro_index.plot(kind='bar',width=0.8, figsize=(10,4),color="pink")
plt.title('host name')
plt.xlabel('nombres repetidos')
plt.ylabel('frecuencia')
```
